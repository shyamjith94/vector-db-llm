```python
# langchain text loader
from langchain.document_loaders import TextLoader
from langchain_community.document_loaders import TextLoader
```

```python
loader=TextLoader("../data/textfile/machine_learning.txt",encoding="utf-8")
loader
```

```python
from pprint import pprint
document = loader.load()
pprint(document)
```

```python
# directly loader
from langchain_community.document_loaders import DirectoryLoader
dir_documents = DirectoryLoader(
    "../data/textfile",
    glob="**/*.txt",
    loader_cls=TextLoader,
    loader_kwargs={"encoding":"utf-8"},
    show_progress=False
)
```

```python
documents = dir_documents.load()
pprint(documents)
```

```python
# pdf loader

from langchain_community.document_loaders import PyMuPDFLoader, PyPDFLoader
```

```python
dir_documents_pdf = DirectoryLoader(
    "../data/textfile",
    glob="**/*.pdf",
    loader_cls=PyMuPDFLoader,
    show_progress=False
)
```

```python
pdf_documents = dir_documents_pdf.load()
pprint(pdf_documents[:3])
```
